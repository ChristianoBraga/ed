Introdução
----------

São sete as notas musicais, cada uma representada por uma letra: dó - C,
ré - D, mi - E, fá - F, sol - G, lá - A, e si - B. *Acordes* são grupos
de notas que obedecem algumas *regras*, definidas sobre *distâncias* em
relação a nota *tônica*.

As distâncias entre duas notas são definidas em termos de *semitons*: um
semitom (ou meio tom) é a menor distância entre duas notas num
instrumento de corda. Em geral, diz-se que a nota meio tom *acima* de
uma nota N dada é N\#, ou N sustenido. A nota meio tom *abaixo* de uma
nota N dada é Nb, ou N bemol.

Todas estas definições podem ser representadas pela seguinte sequência,
que representa a *escala de C maior*:

<a name = "seq">C \| Db/C\# \| D \| Eb/D\# \| E \| F \| Gb/F\# \| G \|
Ab/G\# \| A \| Bb/A\# \| B \| C</a>

onde Db e C\#, por exemplo, representam a mesma nota, assim como Eb e
D\#, Gb e F\#, Ab e G\# e Bb e A\#. Note que não existe E\# ou Fb, ou
B\# e Cb. As notas E e F estão a um semitom de distância, assim como B e
C.

Finalmente, uma *tríade maior* é um acorde dado pela nota tônica, a nota
*dois tons* (terça maior) acima e a que está a *três tons e meio*
(quinta justa) em relação a nota tônica.
